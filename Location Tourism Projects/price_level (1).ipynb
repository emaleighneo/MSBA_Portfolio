```python
merged_df = pd.merge(df_business, df_category, on='business_id', how='inner')

# Now filter the merged DataFrame for Florida restaurants
florida_restaurants = merged_df[
    (merged_df['state'] == 'FL') &
    (merged_df['category_name'] == 'Restaurants')
]
```

```python
# Replace with your actual Yelp API key
api_key = 'gNTGmhlItUBBWuGXY21-wX48h3h4N77JGBDIwDUAcJ-sslUUrnUJlDdT80z0LCZEXm7sNq7LsbGdhv4i4WinakwzIQOTZ1TsPl6dUXlOPQAG2QNbPO2Dnc7r3spMZ3Yx'

# Function to get price levels for a list of restaurant names
def get_price_levels_from_yelp(restaurant_names, cities):
    try:
        client = YelpAPI(api_key)

        # Create an empty list to store restaurant details
        restaurant_details = []

        # Loop through each restaurant name to get the price level
        for index in range(len(restaurant_names)):
            print(f'searching for {restaurant_names[index]}')
            # Perform a Yelp search query for each restaurant name
            results = client.search_query(term=restaurant_names[index], location=f'{cities[index]}, FL, United States', limit=1)

            # Check if any business results are returned
            if results['businesses']:
                business = results['businesses'][0]
                # Extract the necessary details (name, price level)
                restaurant_info = {
                    'name': business.get('name'),
                    'price': business.get('price', 'N/A')  # Default to 'N/A' if price level is not available
                }
            else:
                # If no results found for this restaurant, set 'N/A' for price level
                restaurant_info = {
                    'name': restaurant_names[index],
                    'price': 'N/A'
                }

            # Append the restaurant info to the list
            restaurant_details.append(restaurant_info)

        # Convert the list of restaurant details into a DataFrame
        restaurant_df = pd.DataFrame(restaurant_details)
        return restaurant_df

    except Exception as e:
        print(f"Error occurred while retrieving data: {e}")
        return pd.DataFrame()
```
